### Conversão por unidade

```python
def get_ml(txt):
    ml = txt.find(r'ML')
    if (ml != -1):
        for i in range(4,1,-1):
            v = txt[ml-i:ml]
            if v.isdigit():
                return int(v)
        return 0
    else:
        return 0  
```

```python
notas['ML'] = notas['Produto'].apply(get_ml)
```

```python
notas['valor/ml'] = notas['Valor Unitário']/notas['ML']
```

```python
notas.loc[(notas.Produto.str.find(r'CX') == -1)&(notas.Produto.str.find(r'VINAGRE') == -1)&(notas['ML'] != 0)].groupby('ML')['valor/ml'].describe()
```

# CLASSIFICAÇÃO

```python
notas['vinagre'] = notas['Produto'].apply(lambda x: True if x.find(r'VINAGRE') != -1 else False)
```

```python
notas['vinagre'].value_counts()
```

```python
X = notas['Produto']
y = notas['vinagre']
```

```python
X_train= X[:2500]
X_test= X[2500:]
y_train= y[:2500]
y_test= y[2500:]
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
X_train_counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, y_train)
```

```python
X_test_tfidf = tf_transformer.transform(count_vect.transform(X_test))
y_pred = clf.predict(X_test_tfidf)
```

```python
sum(y_pred == y_test)/len(y_pred)
```
